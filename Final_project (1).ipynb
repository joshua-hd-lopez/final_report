### INTRODUCTION:
Diabetes mellitus is a chronic disease characterized by the body's inability to produce or respond effectively to insulin, resulting in abnormal metabolism of carbohydrates and elevated blood glucose levels. It is one of the most common non-communicable diseases, affecting more than 537 million people globally. There are multiple factors that determine a person’s predisposition to diabetes. Diabetes can cause sever health complications. The ability to accurately diagnose diabetes during the early stages can be challenging but can have significant medical impact on patients.

#### DESCRIPTION OF PAPER:
This paper presents an automatic and comprehensive approach to a diabetes prediction system using the Pima Indian diabetes dataset as well as a dataset sample taken from 203 Bangladesh women.  The methodology involved in this research include data preprocessing, model selection, performance evaluation as well as the development of a mobile application in addition to web deployment. 
*********************************************************************************************************************
#### PROBLEM STATEMENT :
Diagnosing diabetes mellitus during initial development is challenging, the conventional methods of diabetes diagnosis often rely on clinical symptoms, laboratory tests and analyzing multiple patient attributes, which may not always detect the disease at its earliest stages. By leveraging machine learning algorithms and ensemble techniques, this research aims to develop a program capable of accurately identifying individuals at risk of diabetes.
*********************************************************************************************************************
#### CONTEXT OF THE PROBLEM:
Diabetes diagnosis requires analyzing multiple patient attributes, and machine learning allows medical professionals to automate this process. While there have been several other studies on developing diabetes predictions systems, they have mainly focused on accuracy as the sole performance metric. This research expands its evaluation criteria to include precision, recall, and F1 score. By incorporating these additional metrics, the research aims to provide a more comprehensive evaluation of the predictive models developed for diabetes prediction.
*********************************************************************************************************************
#### SOLUTION:
The proposed solution is to build on previous work and develop a machine learning predictions system that has comprehensive evaluation criteria. Building the solution involves the following steps: 

1.	Data Collection and Processing 
    •This study combines the two datasets Pima Indian dataset and an independent dataset of 203 women in Bangladesh (RTML). 
    •Before merging the two datasets, missing insulin values in the RTML dataset had to be addressed. The researchers used a semi supervised learning approach to predict the missing values. This was done by training regression on the Prima Indian dataset to predict insulin levels and apply it to RTML
    • Adaptive Synthetic Sampling (ADASYN) and Synthetic Minority Over-sampling Technique (SMOTE) were used to account for the imbalance between  diabetes and non diabetes samples. 
2.	Model Selection
    •After the data is processed relevant attributes for diabetes prediction are identified using GridSearchCV framework.
3.	Performance Evaluation 
    •Machine learning classifiers including decision trees, random forest, KNN, SVM, logical regression, adaboost, XGBoost, Voting and bagging are used in the evaluation.
4. Deployment 
    • Deployment on web application used HTML and CSS for the frontend and Spyder for the back end. 
    • Deployment on smartphone android was done using andriod studio for front end and Java for coding.
